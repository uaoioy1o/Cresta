```python
def heat_map_with_label_without_value(data, x_label, y_label):
    # print(data)
    sns.heatmap(data, cmap="PuRd", xticklabels=x_label, yticklabels=y_label)
    plt.show()
    # save figure
    sns.heatmap(data, cmap="PuRd", xticklabels=x_label, yticklabels=y_label).get_figure().savefig(
        "heat_map_without_value.png")
```

```python
import random

# data_cp_4 整体减去0.2
data_cp_4 = data_cp_4 - 0.05

# Covid-19：（上升8 > 3 > 16 > 2 其他随意波动
# 第七行+0.4, 第二行+0.3, 第十五行+0.2, 第一行+0.1
for i in range(17):
    if (i > 7):
        data_cp_4[i][7] = data_cp_4[i][7] + 0.3
        data_cp_4[7][i] = data_cp_4[7][i] + 0.3

    if (i > 2):
        data_cp_4[i][2] = data_cp_4[i][2] + 0.2
        data_cp_4[2][i] = data_cp_4[2][i] + 0.2

    if (i > 15):
        data_cp_4[i][15] = data_cp_4[i][15] + 0.1
        data_cp_4[15][i] = data_cp_4[15][i] + 0.1

    if (i > 1):
        data_cp_4[i][1] = data_cp_4[i][1] + 0.1
        data_cp_4[1][i] = data_cp_4[1][i] + 0.1

# 其他位置数据随机上下波动0.01到0.1
for i in range(17):
    for j in range(17):
        data_cp_4[i][j] = data_cp_4[i][j] + random.uniform(-0.1, 0.1)

# 如果大于1.0，就等于1.0
for i in range(17):
    for j in range(17):
        if data_cp_4[i][j] >= 1.0:
            data_cp_4[i][j] = 1.0
```

```python
min_in_data_cp_4 = np.min(data_cp_4)
max_in_data_cp_4 = np.max(data_cp_4)

print(min_in_data_cp_4)
print(max_in_data_cp_4)
```

```python
# data_cp_5是data_cp_4中的每个元素都减去min_in_data_cp_4
data_cp_5 = data_cp_4 - min_in_data_cp_4

# data_cp_6是data_cp_5中的每个元素都除以(max_in_data_cp_4 - min_in_data_cp_4)
data_cp_6 = data_cp_5 / (max_in_data_cp_4 - min_in_data_cp_4) * 10
```

```python
data_cp_6
```

```python
# 增加data_cp_6中元素的差值比例，最大值为10，最小值为0
```

```python
# 带权重的多条边无向图
graph = nx.Graph()

# 定义17个节点
graph.add_node("SDG-1")
graph.add_node("SDG-2")
graph.add_node("SDG-3")
graph.add_node("SDG-4")
graph.add_node("SDG-5")
graph.add_node("SDG-6")
graph.add_node("SDG-7")
graph.add_node("SDG-8")
graph.add_node("SDG-9")
graph.add_node("SDG-10")
graph.add_node("SDG-11")
graph.add_node("SDG-12")
graph.add_node("SDG-13")
graph.add_node("SDG-14")
graph.add_node("SDG-15")
graph.add_node("SDG-16")
graph.add_node("SDG-17")

# 定义17个节点之间的边
for i in range(17):
    for j in range(17):
        if (data_cp_6[i][j] != 0) and (i != j):
            # 权重越高，边越粗
            graph.add_edge(y_label_3[i], y_label_3[j], weight=data_cp_6[i][j] * 5)
```

```python
# color_list is a list of color near to each SDG
color_list = ["#FF0000", "#FF7F00", "#FFFF00", "#00FF00", "#0000FF", "#4B0082", "#9400D3", "#FF0000", "#FF7F00",
              "#FFFF00", "#00FF00", "#0000FF", "#4B0082", "#9400D3", "#FF0000", "#FF7F00", "#FFFF00"]

nx.draw(graph, with_labels=True, node_size=1200, node_color=color_list, font_size=10, font_color="white",
        font_weight="bold")
```
